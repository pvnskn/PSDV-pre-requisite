```python
#1. List Operations and Basic Statistics:

l = [i+1 for i in range(20)]
s = 0
for i in range(10):
    s+=l[2*i+1]
p = 1
for i in range(10):
    p*=l[2*i]
mean = sum(l)/20
median = (l[9]+l[10])/2
var = 0
for i in range(20):
    var+=(l[i]-mean)**2
var/=20
print(f'\nSum of even numbers: {s}\nProduct of odd numbers: {p}')
print(f'\nMean: {mean}\nMedian: {median}\nStandard Deviation: {var**.5}\n')
```

```python
#3. Fibonnaci Sequence:

import matplotlib.pyplot as plt

def fibonnaci(n):
    f = [0,1]
    for i in range(n-2):
        f.append(f[i]+f[i+1])
    return f

n = int(input("Enter a number: "))  #15
x = [i+1 for i in range(n)]
print(f'\nFirst {n} fibonnaci numbers are: {fibonnaci(n)}')

plt.plot(x,fibonnaci(n),color='red')
plt.title('Fibonnaci Numbers')
plt.show()
print()
```

```python
#10. Character Frequency:

import matplotlib.pyplot as plt

def frequency(string):
    log = {}
    for i in range(97,97+26):
        log[chr(i)] = 0
    for c in string:
        log[c]+=1
    return log

string = input('Enter string: ').lower()    
log = frequency(string)
print(f'\nFrequncy of characters in the string:\n{log}\n')
chrs = list(log.keys())
fre = list(log.values())

fig = plt.figure(figsize=(7,5))
keys = list(log.keys())
plt.bar(range(26),fre,tick_label=chrs,color='cyan')
plt.title('Letter Frequency')
plt.show()
print()
```

```python
#13. Sorting Dictionaries:

def dictsort(lst,srtby):
    l = len(lst)
    for i in range(l):
        for j in range(i+1,l):
            if lst[i][srtby] > lst[j][srtby]:
                lst[i],lst[j] = lst[j],lst[i]
    return lst
            
lst = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}, {'name': 'Charlie', 'age': 15}, {'name': 'Daniel', 'age': 40}
      , {'name': 'Emily', 'age': 31}, {'name': 'Fenwick', 'age': 49}]
srtby = input('\nSort the list by the key: ')
print(f'\nThe sorted list by the key "{srtby}": \n{dictsort(lst,srtby)}\n')
```

```python
#16. Set Operations:

from matplotlib_venn import venn2_unweighted
import matplotlib.pyplot as plt

def union(a,b):
    u = a[:]
    for element in b:
        if element not in u:
            u+=[element]
    return sorted(u)

def intersection(a,b):
    i = []
    for element in a:
        if element in b:
            i.append(element)
    return sorted(i)

def symmetric_diff(a,b):
    u = union(a,b)
    i = intersection(a,b)
    sd = []
    for element in u:
        if element not in i:
            sd.append(element)
    return sorted(sd) 

def set_unique(s,sd):
    su = []
    for element in s:
        if element not in sd:
            su.append(element)
    return su

a = [1,2,3,4,7,10]
b = [4,5,6,7,8,9,10]
u = union(a,b)
i = intersection(a,b)
sd = symmetric_diff(a,b)
a_ = set_unique(a,i)
b_ = set_unique(b,i)

print(f'\nSet A: {a}\nSet B: {b}\nUnion of A and B: {u}\nIntersection of A and B: {i}\nSymmetric Diff. of A and B: {sd}\n ')
venn2_unweighted(subsets = (str(a_), str(b_), str(i)), 
                 set_labels = ('Group A',  
                               'Group B'), 
                 set_colors=("orange", 
                             "blue"),alpha=0.7) 
plt.show()
print()
```

```python
#18. Word Count and Frequency:

import matplotlib.pyplot as plt

f = open('text.txt','r')
log = {}
w = 0
print('\nThe text in file:\n')
for line in f.readlines():
    print(line,end='')
    line = line.strip().split()
    for word in line:
        w+=1
        log[word.lower()] = log.get(word.lower(),0)+1

xax = list(log.keys())
yax = list(log.values())
l = len(yax)
word = ''
mx = 0

for i in range(l):
    if yax[i] > mx:
        word = xax[i]
        mx = yax[i]
    elif yax[i] == mx:
        word+=(' '+xax[i])
print(f"\n\nMost frequent word(s): {word}\nIt's frequency: {mx}")

fig = plt.figure(figsize=(11,5))
plt.bar(xax,yax,color = 'yellow')
plt.xlabel('\nWords')
plt.ylabel('Frequency')
plt.title('Frequency of each word in text')
plt.xticks(rotation=90)
plt.show()
print()
```

```python
#21. Basic Statistics:

import random
import matplotlib.pyplot as plt

nums = []
l = random.randint(30,31)
for i in range(l):
    nums.append(random.randint(1,15))
print(f'\nGenerated list: {nums}')
mean = sum(nums)/l

nums.sort()
if l%2 == 0:
    median = (nums[l//2]+nums[l//2-1])/2
else:
    median = nums[l//2]

mode,num = '',0
c,mx = 0,0
for i in range(l):
    if c == 0 or nums[i] != num:
        if c>mx:
            mx = c
            mode = str(num)
        elif c == mx:
            mode+=' '+str(num)
        c = 1
        num = nums[i]
    else:
        c+=1
if c>mx:
        mx = c
        mode = str(num)
elif c == mx:
    mode+=' '+str(num)

var = 0
for num in nums:
    var+=(num-mean)**2
var/=l
sd = var**.5

print(f'\nList after sorting: {nums}\n\nMean: {mean}\nMedian: {median}\nMode,Frequency: {mode},{mx}\nVariance: {var}\nStandard Deviation: {sd}')
print()

fig = plt.figure(figsize=(5,5))
plt.boxplot(nums,bootstrap = 1000)
plt.title('Boxplot of generated data')
plt.show()
```

```python
#22. Normalization:

import random
import matplotlib.pyplot as plt

def Normalize(l):
    mx = max(l)
    mn = min(l)
    n = []
    for num in l:
        n.append((num-mn)/(mx-mn))
    return n

l = random.sample(range(1,100),25)
print(f'\nGenerated list:  {l}')
print(f'\nNormalized list: {Normalize(l)}')

fig = plt.figure(figsize=(5,5))
plt.title('\nNormalized Data')
plt.plot(l,Normalize(l))
plt.show()
print()
```

```python
#25. Leap Year Checker:

def leapyear(n):
    if n%4 == 0 and n%100 != 0:
        return "'a leap year'."
    elif n%400 == 0:
        return "'a leap year'."
    else:
        return "'not a leap year'."

n = int(input("\nEnter a year: "))
print(f'\n{n} is {leapyear(n)}')
```
